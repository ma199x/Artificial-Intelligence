```python
import networkx as nx
import pandas as pd
d={'from':[1,1,2],'to':[2,3,3]}
df=pd.DataFrame(d)
print(df)
```

```python
import networkx as nx
import matplotlib.pyplot as plt

g=nx.from_pandas_edgelist(df,"from","to")
plt.figure(1,figsize=(10,10))
plt.subplot(221)
nx.draw_networkx(g,pos=nx.circular_layout(g),node_color="red")
plt.title("circular",)

plt.subplot(222)
nx.draw_networkx(g,pos=nx.kamada_kawai_layout(g),node_color="green")
plt.subplot(223)

nx.draw_networkx(g,pos=nx.random_layout(g),node_color="blue")
plt.title("random")

plt.subplot(224)
nx.draw_networkx(g,pos=nx.spectral_layout(g),node_color="black")
plt.title("spectral")

plt.show()
```

```python
import networkx as nx
import matplotlib.pyplot as plt
g=nx.Graph()
g.add_node(1,h='a');g.add_node(2,h='b');g.add_node(4,h='c')

g.add_edge(1,2,weight=1);g.add_edge(2,4,weight=3)

pos=nx.spiral_layout(g)

w=nx.get_edge_attributes(g,name="weight")
h=nx.get_node_attributes(g,name='h')
print(h)

nx.draw(g,pos,with_labels=False,font_color="black",node_size=1600)
nx.draw_networkx_edge_labels(g,pos,edge_labels=w,font_size=20,font_color="green")
nx.draw_networkx_labels(g,pos, font_size=20,labels=h, font_color="red", verticalalignment='top')
plt.show()
```
