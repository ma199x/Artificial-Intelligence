```python
# graphName.add_node(one node)
# graphName.add_edge(from_node,to_node)
import networkx as nx
import matplotlib.pyplot as plt
g=nx.Graph()
g.add_node(1)
g.add_node(2)
g.add_node("a")

g.add_edge(1,"a")
g.add_edge(1,2)
g.add_edge(2,"a")
nx.draw(g,with_labels=True)
print(g)
```

```python
# graphName.add_adges_from(list of nodes)
# graphName.add_edges_from(  [(edge,edge),(edge,edge),(edge,edge) ...n ]
import networkx as nx
import matplotlib.pyplot as plt

g=nx.Graph()
# g.add_nodes_from([1,2,3,4,5])
g.add_edges_from(  [(1,2),(2,3),(5,4),(4,2),(1,5)]  )
nx.draw(g,pos=nx.circular_layout(g),with_labels=1,node_color="red",node_size=1500,font_size=20,font_color="blue")
```

```python
import networkx as nx
import matplotlib.pyplot as plt
graph={
    "a":["b","c"],
    "b":["d","e"],
    "c":["f"],
    "e":["f"],
    "d":[],
    "f":[]
}
g=nx.DiGraph()
for k,v in graph.items():
    for new in v:
        g.add_edge(k,new)
        print(k,new) # not required 
nx.draw(g,with_labels=1)

```

```python
gdict={
    "A": ['B', 'C'] ,
    "B": ['A', 'D', 'E'],
    "C": ['A', 'E'] ,
    "D": ['B', 'E', 'F'],
    "E": ['B', 'C', 'D', 'F'],
    "F": ['D', 'E'] 
}
g=nx.Graph()
for k,v in gdict.items():
    for element_v in v:
        g.add_edge(k,element_v)
nx.draw(g)
```
