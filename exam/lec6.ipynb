```python
import networkx as nx
import pandas as pd
import matplotlib.pyplot as plt

d={
    "from":[1,1,2],
    "to":  [2,3,3],
    "ee":[1,2,3]
}
df=pd.DataFrame(d)
print(df)
```

```python
plt.figure(1)
g=nx.from_pandas_edgelist(df,"from","to")
nx.draw_networkx(g)
```

```python
plt.subplot(222)
nx.draw_networkx(g,pos=nx.circular_layout(g))
plt.title("Circular")

plt.subplot(224)
nx.draw_networkx(g,pos=nx.circular_layout(g),node_color='red')


plt.subplot(223)
nx.draw_networkx(g,pos=nx.kamada_kawai_layout(g),node_color='blue')
```

```python
import networkx as nx
import matplotlib.pyplot as plt

g=nx.Graph()
g.add_node(1 , name="one")
g.add_node(2 , name="two")
g.add_node(3 , name="tree")
g.add_node(4 , name="four")

g.add_edge(1,2,d=11)
g.add_edge(1,3,d=12)
g.add_edge(2,3,d=13)
g.add_edge(3,4,d=14)


pos=nx.circular_layout(g)
di=nx.get_edge_attributes(g,"d")
n=nx.get_node_attributes(g,"name")
nx.draw(g,pos,with_labels=0,node_size=1500)
nx.draw_networkx_edge_labels(g,pos,edge_labels=di)
nx.draw_networkx_labels(g,pos,labels=n,verticalalignment="baseline")
# verticalalignment can be ( top', 'bottom', 'center', 'baseline', 'center_baseline)
plt.show()
```
